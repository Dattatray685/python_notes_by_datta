Given a number N, print the sum of the squares of its digits. Input Size : 1 <= N <= 1000000000000000000

Sample Testcase :

INPUT

19

OUTPUT

82

```python
 a = int(input(" enter your num "))
 sum = 0
 while a>0:
  digits = a % 10
  sum += digits * digits
  a = a // 10


print(sum)
```

Write a code to get an integer N and print the sum of values from 1 to N.

Input Description:

A single line contains an integer N.

Output Description:

Print the sum of values from 1 to N.

Sample Input :

10

Sample Output :

55

```python
a = int(input("enter your number :-"))
sum = 0

for i in range (1, a+1):
   sum += i

print(sum)
```

Write a code get an integer number as input and print the sum of the digits.

Input Description:

A single line containing an integer.

Output Description:

Print the sum of the digits of the integer.

Sample Input :

124

Sample Output :

7

```python
a = int(input("enter your num "))
sum = 0
while a>0:
  digits = a % 10
  sum += digits
  a = a // 10


print(sum)
```

Given a number 'N' print the sum of each digit to the power of number of digits in given input.

Example :

Input => 1234 => ( 1 ^ 4 ) + ( 2 ^ 4 ) + ( 3 ^ 4 ) + ( 4 ^ 4 ) => 1 + 16 + 81 + 256 Output => 354

N <=100000000000

Sample Testcase :

INPUT

1234

OUTPUT

354

```python
a = input("enter your num ")
count = len(str(a))
sum = 0
for digits in a:
  sum += int(digits) ** count

print(sum)
```

Given a number N, find the factorial of N.

Input Size : 1 <= N <= 25

Sample Testcase :

INPUT

5

OUTPUT

120

```python
a = int(input("enter your num "))
fact = 1
for i in range(1, a+1):
  fact *= i

print(fact)
```

Given a number N in decimal convert it into binary value.

Input Size : N <= 100000

Sample Testcase :

INPUT

5

OUTPUT

101

```python
n = int(input("enter your num"))
binary = " "

while n > 0:
    binary = str(n % 2) + binary
    n = n // 2

print(binary)
```
